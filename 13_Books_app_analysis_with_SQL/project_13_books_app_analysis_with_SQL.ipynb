### Для каждой книги посчитать количество обзоров и среднюю оценку

```python
query = """ WITH tmp_books_rating AS (
                                      SELECT b.book_id AS book_id,
                                             AVG(r.rating) AS avg_rating
                                      FROM books b LEFT JOIN ratings r on b.book_id=r.book_id
                                      GROUP BY b.book_id
                                     ),
                  tmp_books_review AS (
                                        SELECT b.book_id AS book_id,
                                               COUNT(rev.review_id) AS count_review
                                        FROM books b LEFT JOIN reviews rev on b.book_id=rev.book_id 
                                        GROUP BY b.book_id 
                                      )                   
                                     
            SELECT b.book_id AS book_id,
                   b.title AS title,
                   tmp_books_review.count_review AS count_review,
                   tmp_books_rating.avg_rating AS avg_rating
            FROM books b JOIN tmp_books_review on b.book_id=tmp_books_review.book_id 
                         JOIN tmp_books_rating ON b.book_id=tmp_books_rating.book_id
          
        """
result2 = pd.io.sql.read_sql(query, con=engine)
result2
```

### Определить издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
query = """
            SELECT p.publisher AS publisher_name,
                  COUNT(b.book_id) AS count_books
            FROM books b JOIN publishers p ON b.publisher_id=p.publisher_id
            WHERE b.num_pages > 50
            GROUP BY p.publisher_id
            ORDER BY COUNT(book_id) DESC
            LIMIT 1
        """
result3 = pd.io.sql.read_sql(query, con=engine)
result3
```

### Определить автора с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками

```python
query = """ 
            WITH tmp_books AS (
                               SELECT b.book_id AS book_id
                               FROM books b JOIN ratings r ON b.book_id = r.book_id
                               GROUP BY b.book_id
                               HAVING COUNT(r.rating) >= 50
                              )
                                      
            SELECT a.author AS author_name,
                   AVG(r.rating) AS avg_ratings    
            FROM books b JOIN authors a ON b.author_id = a.author_id
                         JOIN ratings r ON b.book_id = r.book_id
                         JOIN tmp_books tb ON b.book_id = tb.book_id
            GROUP BY a.author_id
            ORDER BY AVG(r.rating) DESC
        """

result4 = pd.io.sql.read_sql(query, con=engine)
result4 
```

**Промежуточные выводы**

Автор с самой высокой средней оценкой книг -  J.K. Rowling/Mary GrandPré с оценкой 4,29. J.K. Rowling автор популярных книг о Гарри Потере, Mary GrandPré - иллюстратор. 
    
Можно обратить внимание, что для некоторых книг в поле автор имеется несколько имен, при этом эти имена не всегда принадлежат авторам, например, в случае с самой высокой средней оценкой книг Mary GrandPré является иллюстратором.
Это следует учитывать в будущем при построении аналитики по авторам.

### Посчитать среднее количество обзоров от пользователей, которые поставили больше 50 оценок

```python
query = """ 
            WITH tmp_users AS (
                               SELECT username
                               FROM ratings
                               GROUP BY username
                               HAVING COUNT(rating_id) > 50
                              )
                              
            SELECT ROUND(AVG(tmp.count_reviews),2) AS avg_reviews
            FROM (
                  SELECT COUNT(r.review_id) AS count_reviews
                  FROM reviews r JOIN tmp_users tu ON r.username = tu.username
                  GROUP BY r.username
                 ) tmp
        """
result5 = pd.io.sql.read_sql(query, con=engine)
result5
```

<div style="border:solid green 3px; padding: 20px">
    
**Промежуточные выводы**

Среднее количество обзоров от пользователей, которые поставили больше 50 оценок, составило 24,3. Активно читающие пользователи оставляют больше обзоров на книги.
