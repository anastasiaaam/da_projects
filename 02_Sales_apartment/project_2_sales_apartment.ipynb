Возможно произошел технический сбой при автоматическом заполнении данных, возникла ошибка при определении геопозиции объектов недвижимости. Попытаемся заполнить пропущенные значения в данных столбцах медианным значением по соответствующему столбцу в разрезе значений `locality_name`, т.е. медианным значением в рамках одного населенного пункта.*

```python
# создадим список с наименованиями столбцов, заполненных автоматически на основе картографических данных
# для каждого такого столбца найдем медианное значение в разрезе населенных пунктов locality_name
# с помощью fillna() заполним пропущенные значения в столбце в разерезе территорий
geo_column_list = ['parks_nearest','ponds_nearest','city_centers_nearest','airports_nearest']
for value in df['locality_name'].unique():
    for column in geo_column_list:
        median_fill = df[(df['locality_name'] == value) & (df[column].isna() == False)][column].median()    
        df.loc[df['locality_name'] == value, column] = df[df['locality_name']== value][column].fillna(median_fill)
```

Попробуем заполнить пропущенные значения медианным значением в разрезе количества этажей в домах, в которых расположены объекты недвижимости.*

```python
# изучим описательную статистику 'ceiling_height'
df['ceiling_height'].describe()
```

```python
# видим, что имеются аномально высокие значения
df.query('ceiling_height > 20')
```

```python
# похоже, что пользователь ошибся с вводом значения высоты потолков 
# вследствие, чего отсутствует разделитель между целой и дробной частью
# разделим значения ceiling_height, которые больше 20, на 10
df['ceiling_height'] = df['ceiling_height'].where((df['ceiling_height']<20),df['ceiling_height']/10)
```

```python
# строим сводную таблицу по исходному датафрейму df по столбцу `floors_total`
# считаем медиану значения в столбце `ceiling_height`, используя метод median()
# результат сохраняем в grouped_data
grouped_data = (df[df['ceiling_height'].isna() == False]
               .pivot_table(index='floors_total', values='ceiling_height', aggfunc='median'))
```

```python
# определим функцию fill_ceiling_height, которая на вход получает строку row
# возвращает по количеству этажей медианное значение ceiling_height из сводной таблицы grouped_data
# возможное отсутствие необходимого количества этажей в grouped_data обработаем конструкцией try ... except ... 
# в случае отсутствия необходимого количества этажей оставим значение row['ceiling_height'] без изменений
def fill_ceiling_height(row):
        try:
            floors = int(row['floors_total'])
            return grouped_data.loc[floors,'ceiling_height']
        except:
            return row['ceiling_height']

# заполним пропущенные значения ceiling_height с помощью метода apply() и функции fill_ceiling_height()
df.loc[df['ceiling_height'].isna() == True,'ceiling_height'] = df.apply(fill_ceiling_height, axis=1)

# оставшиеся пропуски заполним медианным значением по всему столбцу ceiling_height 
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
```

Вероятно, что данные пропуски возникли вследствие того, что пользователь не заполнил указанные столбцы. Попробуем вычислить медианные значения долей `living_area` и `kitchen_area` от `total_area` в разрезе числа комнат и заполнить ими пропуски.*

```python
# создадим отфильтрованный датафрейм с значениями столбцов исходного датафрейма 'rooms','total_area','living_area'
# с непропущенными значениями 'living_area'
df_filtered = df[df['living_area'].isna() == False][['rooms','total_area','living_area']]

# добавим в отфильтрованный датафрейм столбец 'living_area_%',
# значения которого рассчитаем как отношение 'living_area' / 'total_area'
df_filtered['living_area_%'] = df_filtered['living_area'] / df_filtered['total_area']


# строим сводную таблицу по исходному датафрейму df по столбцу `rooms`
# считаем медиану значения в столбце `living_area_%`, используя метод median()
# результат сохраняем в grouped_data
grouped_data = df_filtered.pivot_table(index='rooms', values='living_area_%', aggfunc='median')

# определим функцию fill_living_area, которая на вход получает строку row
# возвращает по количеству комнат значение 'living_area', исходя из 'total_area' и данных по доле жилой площади 
# `living_area_%`из сводной таблицы grouped_data
# возможное отсутствие необходимого количества комнат в grouped_data обработаем конструкцией try ... except ... 
# в случае отсутствия необходимого количества комнат оставим значение row['living_area'] без изменений
def fill_living_area(row):
        try:
            rooms = int(row['rooms'])
            return (grouped_data.loc[rooms,'living_area_%'] * row['total_area'])
        except:
            return row['living_area']

# заполним пропущенные значения living_area с помощью метода apply() и функции fill_living_area()
df.loc[df['living_area'].isna() == True,'living_area'] = df.apply(fill_living_area, axis=1)
```

```python
# создадим отфильтрованный датафрейм с значениями столбцов исходного датафрейма 'rooms','total_area','kitchen_area'
# с непропущенными значениями 'kitchen_area'
df_filtered = df[df['kitchen_area'].isna() == False][['rooms','total_area','kitchen_area']]

# добавим в отфильтрованный датафрейм столбец 'kitchen_area_%',
# значения которого рассчитаем как отношение 'kitchen_area' / 'total_area'
df_filtered['kitchen_area_%'] = df_filtered['kitchen_area'] / df_filtered['total_area']

# строим сводную таблицу по исходному датафрейму df по столбцу `rooms`
# считаем медиану значения в столбце `kitchen_area_%`, используя метод median()
# результат сохраняем в grouped_data
grouped_data = df_filtered.pivot_table(index='rooms', values='kitchen_area_%', aggfunc='median')

# определим функцию fill_kitchen_area, которая на вход получает строку row
# возвращает по количеству комнат значение 'kitchen_area', исходя из 'total_area' и данных по доле кухонной площади 
# `kitchen_area_%`из сводной таблицы grouped_data
# возможное отсутствие необходимого количества комнат в grouped_data обработаем конструкцией try ... except ... 
# в случае отсутствия необходимого количества комнат оставим значение row['kitchen_area'] без изменений
def fill_kitchen_area(row):
        try:
            rooms = int(row['rooms'])
            return (grouped_data.loc[rooms,'kitchen_area_%'] * row['total_area'])
        except:
            return row['kitchen_area']

# заполним пропущенные значения kitchen_area с помощью метода apply() и функции fill_kitchen_area()
df.loc[df['kitchen_area'].isna() == True,'kitchen_area'] = df.apply(fill_kitchen_area, axis=1)
```

Таким образом, обнаружили неявный пропуск в столбце `rooms`. Так как это квартиры-студии, либо квартиры с открытой планировкой, вполне возможно, что они представляют собой единое внутреннее пространство, в этом случае количество комнат можно заполнить `1`, а `kitchen_area` заполнить исходя из медианного значения доли кухонной площади в общей площади для количества комнат равным 1.* 

```python
# заменим значение 0 на 1 в столбце 'rooms'
df['rooms'].replace(0,1, inplace=True)

# найдем медианное значение доли кухонной площади в общей площади для количества комнат равным 1
# создадим отфильтрованный датафрейм со значениями столбцов исходного датафрейма 'total_area','kitchen_area'
# с непропущенными значениями 'kitchen_area' и с 'rooms' равным 1
df_filtered = df[(df['kitchen_area'].isna() == False) & (df['rooms'] == 1)][['total_area','kitchen_area']]

# добавим в отфильтрованный датафрейм столбец 'kitchen_area_%',
# значения которого рассчитаем как отношение 'kitchen_area' / 'total_area'
df_filtered['kitchen_area_%'] = df_filtered['kitchen_area'] / df_filtered['total_area']


# заполним пропуски в 'kitchen_area', умножив медианное значение доли кухонной площади на соответствующее значение 'total_area'
df.loc[df['kitchen_area'].isna() == True, 'kitchen_area'] = (df_filtered['kitchen_area_%'].median() 
                                                             * df.loc[df['kitchen_area'].isna() == True,'total_area'])
```

## Расчёты и добавление результатов в таблицу

*Посчитаем и добавим в датафрейм следующую информацию:*
   
- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.

Переведем значения цены на момент снятия объявления с публикации из рублей в тысячи рублей для удобства анализа.  

```python
# переведем значения цены в тысячи рублей
df['last_price'] = df['last_price'] / 1000

# рассчитаем цену квадратного метра
df['price_per_m2'] = df['last_price'] / df['total_area']

# определим день недели, месяц, год публикации объявления
df['weekday_exposition'] = df['first_day_exposition'].dt.weekday
df['month_exposition'] = df['first_day_exposition'].dt.month
df['year_exposition'] = df['first_day_exposition'].dt.year

# рассчитаем долю жилой и кухонной площадей от общей площади
df['living_area_%'] = df['living_area'] / df['total_area']
df['kitchen_area_%'] = df['kitchen_area'] / df['total_area']
```

```python
# определим функцию floor_group, которая на вход получает строку row
# по столбцам 'floor' и'floors_total' строки row определяет тип этажа [первый, последний, другой]
# floor == 1 - первый этаж
# floor == floors_total - последний этаж
# в остальных случаях - другой этаж

def floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'

# создадим столбец 'floor_group', использую функцию определения типа этажа floor_group  
df['floor_group'] = df.apply(floor_group, axis=1)
```
