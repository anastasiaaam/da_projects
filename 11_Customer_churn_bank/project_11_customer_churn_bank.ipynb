### Предобработка данных

```python
# приведение наименований столбцов змеиному и нижнему регистру
data.columns = data.columns.str.lower()
data.rename(columns = {'userid':'user_id', 'creditcard':'credit_card'}, inplace=True)
data.columns
```

```python
# получение описательной статистики
data.describe().T
```

```python
# проверка на наличие явных дубликатов
print('Количество явных дубликатов: ', data.duplicated().sum())
print('Количество явных дубликатов в столбце \'user_id\': ', data['user_id'].duplicated().sum())
```

```python
# проверка уникальных значений бинарных переменных
print('Уникальные значения в столбце \'credit_card\'', data['credit_card'].unique())
print('Уникальные значения в столбце \'loyalty\'', data['loyalty'].unique())
print('Уникальные значения в столбце \'churn\'', data['churn'].unique())
```

```python
# проверка уникальных значений категориальных переменных
print('Уникальные значения в столбце \'city\'', data['city'].unique())
print('Уникальные значения в столбце \'gender\'', data['gender'].unique())
print('Уникальные значения в столбце \'objects\'', data['objects'].unique())
print('Уникальные значения в столбце \'products\'', data['products'].unique())
```

```python
# посчитаем количество и процент пропущенных значений по столбцам
print('Количество пропущенных значений по столбцам')
display(data.isna().sum().sort_values(ascending=False))
print('Доля пропущенных значений по столбцам в процентах')
display(round(data.isna().sum() / data.shape[0] * 100, 2).sort_values(ascending=False))
```

<div class="alert alert-block alert-info">
<font color='Black'>

Доля пропущенных значений в столбце `balance` составляет около 36%, что довольно существенно и их удаление приведет к существенной потере данных. Исследуем подробнее строки с пропущенными значениями и проанализируем распределения признаков.
    

</font> 
</div>     

```python
# построим гистограммы распределения числовых признаков для строк с пропусками
data[data['balance'].isna()].drop(['balance','user_id'], axis=1).hist(figsize=(15,10), color='#4E944F')
plt.suptitle('Распределения числовых признаков для клиентов с пропусками в признаке <баланс на счете>');
```

```python
# построим гистограммы распределения нечисловых признаков для строк с пропусками
fig, axes = plt.subplots(1, 2, figsize = (12, 4))
plt.suptitle('Распределение нечисловых признаков для клиентов с пропусками в признаке <баланс на счете>');
data[data['balance'].isna()]['city'].value_counts(normalize=True).plot.bar(ax=axes[0])
data[data['balance'].isna()]['gender'].value_counts(normalize=True).plot.bar(ax=axes[1])
plt.show();
```

```python
# сгруппируем строки с пропусками по признакам город, отток, наличие кредитной карты
# и найдем абсолютное и относительно число клиентов для каждой подгруппы 
grouped = (data[data['balance'].isna()]
           .groupby(['city', 'churn', 'credit_card'], as_index=False)['user_id']
           .agg('count'))
grouped['%'] = grouped['user_id'] / grouped['user_id'].sum()
grouped
```

<div class="alert alert-block alert-info">
<font color='Black'>

Видим, что большая часть пропусков в столбце `balance` приходится на оставшихся (не ушедших в отток) клиентов из Ярославля и Рыбинска, которые имеют кредитные карты. Возможно, причина пропусков состоит в отсутствии у клиента текущего счета и, как следствие, остатка на нем, например, клиент пользуется только кредитной картой или пользуется услугами банка, которые не предполагают открытие текущего счета. Оставим пропущенные значения без изменений, так как заполнение их нулевым значением приведет к искажению среднего значения баланса на счете.


</font>

```python
# построим гистограммы распределения числовых признаков для строк с низкими значениями заработной платы
(data[data['estimated_salary'] < 5000].drop(['estimated_salary', 'user_id'], axis=1)
                                     .hist(figsize=(15,10), color='#FFEF82'))
plt.suptitle('Распределение числовых признаков для клиентов с заработной платой < 5000');
```

- медианная заработная плата для женщин немного больше, чем для мужчин (107 000 и 98 000 соответственно).

```python
def graph_for_category(data, column_x, column_y, ax=None, normalize=False):
    
    '''функция для построения гистограммы распределения категориального признака column_x 
       в разрезе категориального признака column_y
       
       параметры функции: 
       data – датафрейм с признаками, 
       column_x – наименование категориального признака, для которого строится гистограмма
       column_y – наименование категориального признака, в разрезе которого строится гистограмма, 
       ax – объект осей для построения гистограммы, значение по умолчанию None, 
       normalize – параметр булевого типа, задающий вид гистограммы, 
                   если значение True, то гистограмма будет построена для относительных долей признака column_y , 
                   если значение False, то гистограмма будет построена для абсолютных значений признака column_y , 
                   значение по умолчанию False
                   
       Функция возвращает объект с построенной гистограммой'''
    
    x,y = column_x, column_y
    name_dict = {True: 'доля,%', False: 'количество'}

    hue_order = np.sort(data[column_y].unique())
    if normalize==True:
        df = data.groupby(x)[y].value_counts(normalize=True).mul(100).rename(name_dict[normalize]).reset_index()
    else:
        df = data.groupby(x)[y].value_counts().rename(name_dict[normalize]).reset_index()

    g = sns.barplot(x=x, y=name_dict[normalize], hue=y, hue_order=hue_order, data=df, ax=ax, palette='coolwarm_r')
    
    if normalize==True:
        g.set_ylim(0,100)
        
    g.set_xlabel(category_name_dict[column_x])

    g.legend(title=category_name_dict[column_y])

    if (column_x != 'objects') and (normalize==True):
        for p in g.patches:
            txt = str(p.get_height().round(1)) + '%'
            txt_x = p.get_x() 
            txt_y = p.get_height()
            g.text(txt_x+.05, txt_y+1, txt) 

    return g     
```

## Проверка статистических гипотез

<div class="alert alert-block alert-info">
<font color='Black'>

По условию задачи проверим гипотезу о различии дохода между теми клиентами, которые ушли и теми, которые остались.
    
По результатам исследовательского анализа данных мы выявили, что основными признаками, потенциально влияющими на отток, является возраст и активность клиента. Поэтому также проверим  гипотезу о различии возраста и гипотезу о различии доли активных клиентов между теми клиентами, которые ушли и теми, которые остались.
    
Для проверки гипотез о различии средних значений дохода и возраста применим t-критерий.
    
Для применения t-критерия: 
- выборки не должны зависеть друг от друга, это условие выполняется (в нашем датасете все клиенты уникальные),
- выборочные средние должны быть нормально распределены, это условие выполняется благодаря ЦПТ: если размер выборки составляет хотя бы несколько десятков значений, выборочные средние, которые можно получить из одной и той же генеральной совокупности, будут распределены нормально вокруг истинного среднего этой совокупности.

В случае сравнения выборок разного размера или при наличии сомнений относительно равенства дисперсий рассматриваемых совокупностей, в параметрах t-критерия имеется возможность учесть данные обстоятельства и задать параметр equal_var как False.

Для проверки гипотез о различии доли активных клиентов применим z-критерий для пропорций.
    
</font> 
</div>   

```python
print('Размер выборки оставшихся клиентов:', len(data[data['churn']==0]['user_id']))
print('Размер выборки клиентов, ушедших в отток:', len(data[data['churn']==1]['user_id']))
```

### Проверка гипотезы различия дохода между теми клиентами, которые ушли и теми, которые остались

<div class="alert alert-block alert-info">
<font color='Black'>

Сформулируем нулевую и альтернативную гипотезы.
    
Нулевая гипотеза ($H_0$): средняя заработная плата оставшихся клиентов и средняя заработная плата ушедших в отток клиентов равны.

Альтернативная гипотеза ($H_1$): средняя заработная плата оставшихся клиентов и средняя заработная плата ушедших в отток клиентов различаются.

Построим график распределения заработной платы для оставшихся клиентов и для клиентов, ушедших в отток.
    
</font> 
</div>   

```python
plt.figure(figsize=(12,6))
plt.title('Распределение признака <{}>'.format(numeric_name_dict['estimated_salary']), y=1, fontsize=14)
    
sns.histplot(x=data['estimated_salary'].dropna(), hue='churn', kde=True, data=data, 
                 palette=[colors[0], colors[1]])

plt.xlabel(numeric_name_dict['estimated_salary'])
plt.ylabel('Количество клиентов')
plt.ylim((0,500))
plt.legend(['ушедшие в отток', 'оставшиеся'], title='Признак оттока клиентов', loc='upper right')
    
plt.show()
```

<div class="alert alert-block alert-info">
<font color='Black'>

Анализируя график, можно предположить, что распределения заработной платы для оставшихся клиентов и для клиентов, ушедших в отток, практически одинаковы. 

Проведем  t-тест, критический уровень значимости установим на уровне 5%.
    
</font> 
</div>   

```python
alpha = .05 
# критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(data[data['churn']==0]['estimated_salary'], 
                       data[data['churn']==1]['estimated_salary'],
                       equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

```python
print('Среднее значение заработной платы для выборки оставшихся клиентов:', 
       round(np.mean(data[data['churn']==0]['estimated_salary']),1))
print('Среднее значение заработной платы для выборки клиентов, ушедших в отток:', 
       round(np.mean(data[data['churn']==1]['estimated_salary']),1))
print('Различие средних значений заработной платы в процентах: {:.2%}'
       .format((np.mean(data[data['churn']==0]['estimated_salary']) / 
                np.mean(data[data['churn']==1]['estimated_salary'])) - 1))
```

<div class="alert alert-block alert-info">
<font color='Black'>

Результаты t-теста показывают, что у нас нет оснований для того, что бы отвергнуть нулевую гипотезу, значит различие между средней заработной платой оставшихся клиентов и клиентов, ушедших в отток, не является статистически значимым.
    
</font> 
</div>
